## Investigation Overview


> The analysis of Borrower Rate and Consumer Income was derived from Prosper Loan, we extracted the necessary columns for exploration and no tidiness were conducted in the analysis. The graphical presentation of DSR implies that borrowers with high DSR rate have lower access to loan amount because they are considered high risk for investors. Also, a higher number of employed and full-time employment status placed against loan original amount shows a direct posisitive relations, higher BorrowerAPR is attributed to Other and Not employed employment status which makes it high risk. In conclusion there is a negative correlation between borrrower APR and Loan amount which is highlighted with a downward straightline slope

## Dataset Overview

> There are total of 113937 records and 15 features in our loan datasets used for the prosper loan data analysis.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe

df = pd.read_csv("prosperLoanData.csv")
```

## APR Rating Across Loan Amount Disbursed

> The regplot below shows the relationship between borrrower APR and Loan amount disbursed to customer, a negative correlation exist between the indices because the higher the APR rate the lesser the loan amount disbursed to customers


```python
plt.figure(figsize = [10, 4])
sns.regplot(data = df, x = 'LoanOriginalAmount', y = 'BorrowerAPR', scatter_kws={'alpha':0.10});
plt.title('APR Rating Across Loan Amount Disbursed');
```

## Distribution of Borrower Debt to Income Ratio

> The graph is slightly skewed to the right with a continous downward slope, the highest count of debt income ratio is identified around 0.2, this implies that borrowers with high DSR ratio have lower access to loan because they considered high risk for investors

```python
# distribution of debt income ratio
plt.figure(figsize = [10,5])
sns.distplot(df.DebtToIncomeRatio, bins = 120,kde = False)
plt.xlim(-.1, 1)
plt.xlabel('DebtToIncomeRatio')
plt.ylabel('Count')
plt.title('Distribution of Borrower Debt to Income Ratio');
```

## Monthly loan payment analysed with monthly income

> The plot depicts the relationship of the variables as inversely correlated, it was identified that custoemrs with lower monthly income pays higher monthly loan payments and vice versa 

```python
plt.figure(figsize=[8,4])
sns.regplot(data=df, x='StatedMonthlyIncome', y= 'MonthlyLoanPayment');
plt.title('Monthly loan payment accross stated monthly income');
```
